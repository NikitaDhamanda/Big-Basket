#### Finding Outliers 

```python
sns.boxplot(df['sale_price'])
```

```python
q1=df['sale_price'].quantile(0.25)
q3=df['sale_price'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers=df[(df['sale_price']<lower_bound)|(df['sale_price']>upper_bound)]
outliers
```

```python
mean_value=df['sale_price'].mean()
df['sale_price']=df['sale_price'].apply(lambda x:mean_value if x<lower_bound or x>upper_bound else x)
print(mean_value)
```

```python
median_value=df['sale_price'].median()
df['sale_price']=df['sale_price'].apply(lambda x:mean_value if x<lower_bound or x>upper_bound else x)
print(median_value)
```
